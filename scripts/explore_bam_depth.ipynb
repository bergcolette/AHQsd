```python
import pandas as pd
import numpy as np
import glob
import os
from matplotlib import pyplot as plot
```

```python
dataDir="/home/colette_berg/YNP/AHQsd_2021/bams"


depth_list = glob.glob(os.path.join(dataDir, '*.depth'))
```

```python
# medianFile=pd.DataFrame(columns=['individual','median_depth'])
fileNames=[]
median=[]

for i in depth_list:
    print(i)
    fileNames.append(i)
    median.append(np.median(pd.read_csv(i, sep="\t", index_col=False, header=None)[2]))
```

```python
summaryFile=pd.DataFrame(columns=['file','median_depth'])
summaryFile['file']=fileNames
summaryFile['median_depth']=median
```

```python
plot.plot(summaryFile['file'], summaryFile['median_depth'])
```
